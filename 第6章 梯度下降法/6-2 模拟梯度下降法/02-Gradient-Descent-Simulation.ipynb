## 梯度下降模拟

```python
import numpy as np 
# import matlplotlib.pyplot as plt
import matplotlib.pyplot as plt
```

```python
plot_x = np.linspace(-1, 6, 141) # 包含了1 和 6，141个点
plot_x # 绘图要用的 x 点
```

```python
plot_y = (plot_x - 2.5)**2-1 # 二次函数：损失函数，找到这样一个曲线的最小值对应的横轴的值
# 中心点向右移动
# 对于这个损失函数是否有原来的模型，有可能没有；但也可能有。关键是没有意义。
# 这一小节的重点是如何用梯度下降法寻找一个函数的最小值，和模型什么的一点关系都没有。
# 我觉得重点是这里只强调了J是theta的1元函数，后面都是对theta的2元以上的函数了。
```

```python
plt.plot(plot_x, plot_y) # 这里模拟了一个一元损失函数
plt.xlabel("theta") # 这里只有一个theta，而对于y=ax+b我们也至少哟两个theta要求，所以对于线性回归模型肯定是梯度了
plt.ylabel("J")
plt.show()
```

```python
def dJ(theta):  # 对J求导，二次方程对J求导的值
    return 2*(theta-2.5)
```

```python
def J(theta): # theta值对应的损失函数是多少？
    print(theta)
    return (theta-2.5)**2-1
```

```python
# eta = 0.01
eta = 0.1
epsilon = 1e-8

theta_initial = 0.0
theta = theta_initial
while True:
    gradient = dJ(theta)
    last_theta = theta
    # 梯度方向是J增大的方向，就得到了 theta 的变化后的值
    theta = theta - eta * gradient
    
    if(abs(J(theta) - J(last_theta)) < epsilon):
        break

print(theta)
print(J(theta))
```

### 添加theta历史值，把中间过程写出来

```python
eta = 0.1
epsilon = 1e-8

theta_initial = 0.0
theta = theta_initial
theta_history = [theta]
while True:
    gradient = dJ(theta)
    last_theta = theta
    # 梯度方向是J增大的方向，就得到了 theta 的变化后的值
    theta = theta - eta * gradient
    theta_history.append(theta)
    
    if(abs(J(theta) - J(last_theta)) < epsilon):
        break
        
# 注意这里的theta_history是python里面的列表，所以必须转化为np.array，因为都是针对numpy来计算的
# plt.plot(theta, J(theta))
plt.plot(plot_x, J(plot_x)) # theta是曲线的独立参数
plt.plot(np.array(theta_history), J(np.array(theta_history)), color = 'r', marker = '+')
# plt.plot(np.array(theta_history), J(np.array(theta_history)), color=\"r\", marker='+')
plt.xlabel("theta")
plt.ylabel("J")
# plt.show
plt.show() 
# 刚开始挪动的幅度比较大，比较陡，斜率比较大，到后面梯度变得平缓了
```

### 对上面的步骤封装为梯度下降函数

```python
# def gradient_descent(initial_theta = 0.0, theta_history = [], eta = 0.1):
def gradient_descent(initial_theta, eta, epsilon=1e-8): # 传递进去的应该是变量
    theta = initial_theta
    theta_history.append(theta)
    while True:
        gradient = dJ(theta)
        last_theta = theta
        # 梯度方向是J增大的方向，就得到了 theta 的变化后的值
        theta = theta - eta * gradient
        theta_history.append(theta)

        if(abs(J(theta) - J(last_theta)) < epsilon):
            break
            
def plot_theta_history():
    plt.plot(plot_x, J(plot_x)) # theta是曲线的独立参数
    plt.plot(np.array(theta_history), J(np.array(theta_history)), color = 'r', marker = '+')
    plt.xlabel("theta")
    plt.ylabel("J")
    plt.show() 
```

```python
def gradient_descent(initial_theta, eta, n_iters = 1e4, epsilon=1e-8): # 传递进去的应该是变量
    theta = initial_theta
    theta_history.append(theta)
    
    i_iter = 0
    # while True:
    while i_iter < n_iters:
        gradient = dJ(theta)
        last_theta = theta
        # 梯度方向是J增大的方向，就得到了 theta 的变化后的值
        theta = theta - eta * gradient
        theta_history.append(theta) # 越来越大了

        if(abs(J(theta) - J(last_theta)) < epsilon):
            break
        i_iter += 1
```
